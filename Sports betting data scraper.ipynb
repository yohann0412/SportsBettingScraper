```python
import time
import pprint
from selenium import webdriver  # importing seleniums webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.action_chains import ActionChains  # need actionchains to perform advanced functions like click

start = time.time()

driver = webdriver.Chrome()  # Setting up the web driver to run on chrome
driver.get('https://www.nba.com/stats/players/traditional')

# print(driver.find_element("xpath",'/html/body/div[1]/div[2]/div[2]/div[3]/section[2]/div/div[2]/div[3]/table/tbody/tr[1]').text)
selector = driver.find_element("xpath",'/html/body/div[1]/div[2]/div[2]/div[3]/section[2]/div/div[2]/div[2]/div[1]/div[3]/div/label/div/select')
select = Select(selector)

# You can select an option by its index. Indexes are zero-based.
select.select_by_index(0)

# PURPOSE: LOOP THROUGH PLAYERS AND ASSIGN THE DATA
playerData = []

# Common suffixes that might appear in player names
known_suffixes = ['Jr.', 'Sr.', 'III', 'IV', 'II', 'V']

for i in range(563):
    rawData = driver.find_element("xpath", f'/html/body/div[1]/div[2]/div[2]/div[3]/section[2]/div/div[2]/div[3]/table/tbody/tr[{i+1}]').text.split()
    player = rawData[1::]
    
    # Check if the third element is a suffix or not numeric and the fourth element (team) is uppercase (indicating it's not part of the name)
    if not player[2].isnumeric() and (player[2] in known_suffixes or player[3].isupper()):
        name = f'{player[0]} {player[1]} {player[2]}'
        start_index = 3
    else:
        name = f'{player[0]} {player[1]}'
        start_index = 2

    playerData.append({
        "Name": name,
        "Team": player[start_index],
        "Age": player[start_index + 1],
        "GamesPlayed": player[start_index + 2],
        "Wins": player[start_index + 3],
        "Losses": player[start_index + 4],
        "MinutesPlayed": player[start_index + 5],
        "Points": player[start_index + 6],
        "FieldGoalsMade": player[start_index + 7],
        "FieldGoalsAttempted": player[start_index + 8],
        "FieldGoalPercentage": player[start_index + 9],
        "ThreePointersMade": player[start_index+10],
        "ThreePointersAttempted": player[start_index + 11],
        "ThreePointersPercentageMade": player[start_index + 12],
        "FreeThrowsMade": player[start_index + 13],
        "FreeThrowsAttempted": player[start_index + 14],
        "FreeThrowsPercentage": player[start_index + 15],
        "OffensiveRebounds": player[start_index + 16],
        "DefensiveRebounds": player[start_index + 17],
        "TotalRebounds": player[start_index + 18],
        "Assists": player[start_index + 19],
        "Turnovers": player[start_index + 20],
        "Steals": player[start_index + 21],
        "Blocks": player[start_index + 22],
        "PersonalFouls": player[start_index + 23],
        "FantasyPoints": player[start_index + 24],
        "DoubleDouble": player[start_index + 25],
        "TripleDouble": player[start_index + 26],
        "PlusMinus": player[start_index + 27]
    })

    # print(f'Time taken for player {i}: {end - start}')
end = time.time()
print(f'Time taken: {end-start} seconds')
pprint.pp(playerData, sort_dicts=False)
```

```python
import csv

size = len(playerData)
timeTakenForCSV = 0
startTime = time.time()
with open('playerstats.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    fields = [ "Name", "Team", "Age",
             "GamesPlayed", "Wins", "Losses", 
             "MinutesPlayed", "Points", "FieldGoalsMade", 
             "FieldGoalsAttempted","FieldGoalPercentage",
            "ThreePointersAttempted", "ThreePointersPercentageMade", "FreeThrowsMade",
            "FreeThrowsAttempted", "FreeThrowsPercentage", "OffensiveRebounds",
            "DefensiveRebounds", "TotalRebounds", "Assists",
            "Turnovers", "Steals", "Blocks",
            "PersonalFouls", "FantasyPoints", "DoubleDouble",
            "TripleDouble"]
    writer.writerow(fields)
    for data in playerData:
        row = [data[field] for field in fields]
        writer.writerow(row)
    endTime = time.time()
    timeTakenForCSV = endTime - startTime
print("DOWNLOADED CSV")
print(f"Time taken: {timeTakenForCSV}")
```
